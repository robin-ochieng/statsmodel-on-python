# Holt winters method.

```python
airline.head()
```

```python
df1 = airline[["Thousands of Passengers"]]
```

```python
df1.index = pd.to_datetime(df1.index)
```

```python
df1.index
```

```python
df1.index.freq="MS"
```

```python
df1.head()
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing
```

```python
span = 12
alpha = 2/(span+1)
```

```python
df1["EMWA12"]  = df1["Thousands of Passengers"].ewm(alpha=alpha, adjust =False).mean()
```

```python
df1.head()
```

```python
model = SimpleExpSmoothing(df1["Thousands of Passengers"])
```

```python
fitted_model = model.fit(smoothing_level=alpha,optimized=False)
```

```python
df1["SES12"]= fitted_model.fittedvalues.shift(-1)
```

```python
df1.head()
```

```python
df1[["Thousands of Passengers", "EMWA12"]].plot()
```

```python
df2 = df1[["Thousands of Passengers"]]
```

```python
df2.head()
```

```python
df2.index
```

```python
df1["DES__12"] =ExponentialSmoothing(df1["Thousands of Passengers"], trend ="add").fit().fittedvalues.shift(-1)
```

```python
df1.columns
```

```python
df1[['Thousands of Passengers','SES12', 'DES__12']].plot()
```

```python
df1["DES_mat_12"] =ExponentialSmoothing(df1["Thousands of Passengers"], trend ="mul").fit().fittedvalues.shift(-1)
```

```python
df1[['Thousands of Passengers', "DES_mat_12",'SES12', 'DES__12']].plot()
```

```python
df1["TES_MUL_12"] =ExponentialSmoothing(df1["Thousands of Passengers"],trend="mul",seasonal="mul", seasonal_periods =12).fit().fittedvalues
```

```python
model2 = ExponentialSmoothing(df1["Thousands of Passengers"],trend="mul",seasonal="mul", seasonal_periods =12)
```

```python
model2fit = model2.fit()
```

```python
df1["TES_MUL_12"] = model2fit.fittedvalues
```

```python
df1.plot()
```

```python
df1[['Thousands of Passengers', "DES_mat_12","TES_MUL_12"]].plot()
```
